```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import os
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
df_linear_before = pd.read_csv("/home/mddo/stage/M2S4/output/Sigma/accuracy/test/accuracy_HFI_NI_PROM_linear_0_full.csv")
df_linear_before.columns = ["forest","accuracy","precision","recall","fscore","total_tree"]

df_linear_after = pd.read_csv("/home/mddo/stage/M2S4/output/Sigma/accuracy/test/accuracy_HFI_NI_PROM_linear_removed_0.csv")
df_linear_after.columns = ["forest","accuracy","precision","recall","fscore","total_tree"]
```

```python
concat_acc_df = pd.DataFrame()
concat_acc_df["accuracy_before"] = df_linear_before["accuracy"]
concat_acc_df["accuracy_after"] = df_linear_after["accuracy"]

concat_precision_df = pd.DataFrame()
concat_precision_df["precision_before"] = df_linear_before["precision"]
concat_precision_df["precision_after"] = df_linear_after["precision"]
```

```python
ax = plt.gca()
sns.set_style("whitegrid")
concat_acc_df.plot(kind='line',ax=ax)
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(18.5, 10.5)
plt.rcParams["figure.figsize"] = (20,2)
plt.xlabel("Training times", size = 20)
plt.ylabel("Accuracy", size = 20)
plt.xticks(size = 16)
plt.yticks(size = 16)
```

```python
ax = plt.gca()
sns.set_style("whitegrid")
concat_precision_df.plot(kind='line',ax=ax)
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(18.5, 10.5)
plt.rcParams["figure.figsize"] = (20,2)
plt.xlabel("Training times", size = 20)
plt.ylabel("Precision", size = 20)
plt.xticks(size = 16)
plt.yticks(size = 16)
```
